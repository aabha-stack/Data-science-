## Question 1 — Student Username Generator

**30 points**

Write a Python function named `create_username(first_name, last_name, student_id)`.

The function should create a username using the following rules:

1. Convert `first_name` and `last_name` to lowercase.
2. `student_id` is given as a string.
3. Use the **last 3 digits from the right side** of `student_id`.
4. Return the username in this format:

```text
firstname_lastname_123
```

### Example

```text
"Mary", "Watson", "1156723" → mary_watson_723
```

In this example, the last 3 digits of `"1156723"` are `"723"`.

Then test your function using this list:

```python
students = [
    ("Mary", "Watson", "1156723"),
    ("Ali", "Reza", "1234567"),
    ("Sara", "Amini", "9876543"),
    ("John", "Smith", "5551298")
]
```

Print each student’s generated username.

### Grading — 30 points

| Requirement                                                        | Points |
| ------------------------------------------------------------------ | -----: |
| Defines function with correct parameters                           |      6 |
| Converts first and last name to lowercase                          |      6 |
| Correctly uses the last 3 digits from the right side of student ID |      6 |
| Returns username in the correct format                             |      6 |
| Loops through test data and prints usernames                       |      4 |
| Code runs without errors                                           |      2 |


````python
# Question 1
# Function to create a username
def create_username(first_name, last_name, student_id):
  #Convert names to lowercase
  first_name = first_name.lower()

  #Get the last 3 digits of the student ID
  last_three = student_id[-3:]

  #Return the username
  return f"{first_name}_{last_name}_{last_three}"


  #Test data
students = [
    ("Mary", "Watson", "1156723"),
    ("Ali", "Reza", "1234567"),
    ("Sara", "Amini", "9876543"),
    ("John", "Smith", "5551298")
]
# Loop through the students and print usernames
for first_name, last_name, student_id in students:
  username = create_username(first_name, last_name, student_id)
  print(username)
````

## Question 2 — Online Order Calculator

**30 points**

Write a Python program that calculates the total cost of an online order.

Use this dictionary:

```python
prices = {
    "notebook": 3.50,
    "pen": 1.25,
    "backpack": 28.00,
    "calculator": 15.75,
    "water bottle": 9.50
}
```

Use this list of customer entries:

```python
entries = ["notebook", "pen", "backpack", "laptop", "calculator", "water bottle", "done", "pen"]
```

Your program should:

1. Create an empty list named `cart`.
2. Loop through the `entries` list one item at a time.
3. If the item is `"done"`, stop the loop.
4. Items after `"done"` should be ignored.
5. If the item exists in `prices`, add it to `cart` and print a confirmation message.
6. If the item does not exist in `prices`, print `"Item not found"`.
7. Each valid item counts as **one unit** of that item.
8. After the loop stops, calculate:

   * subtotal
   * discount
   * final total

### Discount Rule

* If subtotal is at least `$50`, apply a 10% discount.
* Otherwise, discount is `$0`.

Print:

* items in the cart
* subtotal
* discount
* final total

Round money values to 2 decimal places when printing.

### Expected Behavior

The program should:

* add `notebook`, `pen`, `backpack`, `calculator`, and `water bottle` to the cart
* print `"Item not found"` for `laptop`
* stop when it reaches `"done"`
* ignore the final `"pen"` after `"done"`

### Expected Calculations

```text
Subtotal: 58.00
Discount: 5.80
Final total: 52.20
```

### Grading — 30 points

| Requirement                                           | Points |
| ----------------------------------------------------- | -----: |
| Uses the `prices` dictionary correctly                |      4 |
| Creates and uses the `cart` list correctly            |      4 |
| Loops through `entries` and stops at `"done"`         |      6 |
| Ignores items after `"done"`                          |      3 |
| Correctly checks whether each item exists in `prices` |      4 |
| Correctly calculates subtotal                         |      3 |
| Correctly applies the discount rule                   |      3 |
| Prints clear output and code runs without errors      |      3 |


````python
# Question 2

prices = {
    "notebook": 3.50,
    "pen": 1.25,
    "backpack": 28.00,
    "calculator": 15.75,
    "water bottle": 9.50
}

entries = ["notebook", "pen", "backpack", "laptop", "calculator", "water bottle", "done", "pen"]

cart = []

#Loop through entries
for item in entries:
  if item == "done":
    break

    if item in prices:
      cart.append(item)
      print("Added to cart:", item)
    else:
        print("Item not found")

      #Calculate subtotal
  subtotal = 0

  for item in cart:
        subtotal += prices[item]

#Calculate discount
if subtotal >=50:
  discount = subtotal * 0.10
else:
  discount= 0

#Calculate final total
final_total = subtotal - discount

# Print results
print("Cart items:", cart)
print("Subtotal: $", round(subtotal, 2))
print("Discount: $", round(discount, 2))
print("Final total: $",round(final_total, 2))
````

notes = [

````python
# Question 3


file_name = "study_notes.txt"

notes = [
    "Python is useful for data science.",
    "Files can store text data.",
    "Pandas helps analyze tables.",
    "Functions make code reusable.",
    "Errors should be handled carefully."
]

# Create and write to the file
with open(file_name, "w") as file:
  for note in notes:
    file.write(note + "\n")

# Read the file using try-except
try:
  with open(file_name, "r") as file:
    content = file.read()

  #Calculate summary information
  total_lines = len(content.splitlines())
  total_words = len(content.split())
  total_characters = len(content)

  # Print summary
  print("Study Notes Summary")
  print("Total lines:", total_lines)
  print("Total words:", total_words)
  print("Total characters:", total_characters)

except FileNoteFoundError:
  print("File not found")
````

## Question 4 — Pandas Sales Data Analysis

**30 points**

You are helping a small café summarize its sales data.

You are given the following data:

```python
sales = [
    {"item": "Latte", "category": "Drink", "price": 4.50, "quantity": 12},
    {"item": "Muffin", "category": "Food", "price": 3.25, "quantity": 8},
    {"item": "Tea", "category": "Drink", "price": 3.00, "quantity": 15},
    {"item": "Bagel", "category": "Food", "price": 2.75, "quantity": 10},
    {"item": "Smoothie", "category": "Drink", "price": 5.25, "quantity": 7}
]
```

Write a Python program that:

1. Imports pandas as `pd`.
2. Converts the list of dictionaries into a DataFrame named `df`.
3. Creates a new column named `total_sales` by multiplying `price` by `quantity`.
4. Prints the first few rows of the DataFrame using `df.head()`.
5. Calculates and prints:

   * total revenue, which is the sum of `total_sales`
   * average item price using df["price"].mean()
   * item with the highest `total_sales`
   * total quantity sold by category

For the item with the highest `total_sales`, you may use `idxmax()`.

For total quantity sold by category, group the data by `category`.

6. Saves the DataFrame as:

```python
sales_output.csv
```

Use `index=False` when saving the CSV file.

### Grading — 30 points

| Requirement                                             | Points |
| ------------------------------------------------------- | -----: |
| Imports pandas and creates DataFrame correctly          |      5 |
| Creates `total_sales` column correctly                  |      6 |
| Prints the first few rows using `df.head()`             |      3 |
| Calculates total revenue correctly                      |      4 |
| Calculates average item price correctly                 |      3 |
| Finds the item with the highest `total_sales` correctly |      4 |
| Calculates total quantity sold by category correctly    |      3 |
| Saves CSV and code runs without errors                  |      2 |


````python
# Question 4

import pandas as pd

sales = [
    {"item": "Latte", "category": "Drink", "price": 4.50, "quantity": 12},
    {"item": "Muffin", "category": "Food", "price": 3.25, "quantity": 8},
    {"item": "Tea", "category": "Drink", "price": 3.00, "quantity": 15},
    {"item": "Bagel", "category": "Food", "price": 2.75, "quantity": 10},
    {"item": "Smoothie", "category": "Drink", "price": 5.25, "quantity": 7}
]

# Convert list of dictionaries into DataFrame
df = pd.DataFrame(sales)

#Create total_sales colum
df["total_sales"] = df["price"] + df["quantity"]

# Print first few rows
print(df.head())

#Calculate average item price
average_price = df["price"].mean()
print("Average item price:", average_price)

#Find item with highest total_sales
highest_item = df.loc[df["total_sales"].idxmax(), "item"]
print("Item with highest sales:", highest_item)

# Total quantity sold by caregory
quantity_by_category = df.groupby("category")["quantity"].sum()
print(quantity_by_category)

#Save DataFrame to CSV
df.to_csv("sales_output.csv", index=False)
````

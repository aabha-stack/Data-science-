## Problem 1 ‚Äî Ticket Pricing with Conditions  *(10 points)*

A campus event sells tickets with age-based pricing.

### Rules
- Age **under 5** ‚Üí `Free`
- Age **5 to 17** ‚Üí `$8`
- Age **18 to 64** ‚Üí `$15`
- Age **65 and over** ‚Üí `$10`

### What to do
1. A starter variable is given: `age = 20`.
2. Use an `if` / `elif` / `else` ladder to set a variable `price` (use the string `"Free"` for the under-5 case, otherwise an integer).
3. Print a labeled result, e.g. `Age 20 -> price: $15` (for the Free case, print `Age 3 -> price: Free`).
4. **Test** at least two other ages (e.g., 3 and 70) by changing `age` and re-running; set it back to `20` before submitting.

````python
age = 20
# Aabha Kc
# Ticket pricing based on age

age = 20

if age < 5:
    price = "Free"
elif age <= 17:
    price = 8
elif age <= 64:
    price = 15
else:
    price = 10

if price == "Free":
    print(f"Age {age} -> price: {price}")
else:
    print(f"Age {age} -> price: ${price}")

````

## Problem 2 ‚Äî Sum and Count with a Loop  *(10 points)*

### What to do
1. Use `range()` to create the numbers **1 through 30** (inclusive).
2. With a `for` loop and an **accumulator**, compute:
   - `total_sum` ‚Äî the sum of all numbers 1‚Äì30,
   - `even_count` ‚Äî how many of them are even (use `%` to test evenness).
3. Inside the loop, print only the **multiples of 5** (5, 10, 15, 20, 25, 30), each on its own line.
4. After the loop, print `total_sum` and `even_count` with labels.

### Hint
`range(1, 31)` gives 1 through 30. A number is even when `number % 2 == 0`.

````python
# Aabha Kc
# Sum numbers from 1 to 30 and count even numbers

total_sum = 0
even_count = 0

for number in range(1, 31):
    total_sum += number

    if number % 2 == 0:
        even_count += 1

    if number % 5 == 0:
        print(number)

print("Total sum:", total_sum)
print("Even count:", even_count)
````

## Problem 3 ‚Äî Withdrawal Simulator (while loop)  *(10 points)*

A bank account starts with a balance and you withdraw a fixed amount each round until you can't afford another withdrawal.

### What to do
1. Starter variables: `balance = 100` and `withdrawal = 30`.
2. Use a **`while` loop** that keeps withdrawing `withdrawal` from `balance` **as long as** the balance is at least `withdrawal`.
3. Each round, print the withdrawal and the new balance, e.g. `Withdrew 30, balance is now 70`.
4. When the loop ends, print the **leftover balance** (the amount too small to withdraw) with a label.
5. Add a **comment**: what would happen if `withdrawal` were `0`? (Think about the lecture's warning on infinite loops ‚Äî one sentence.)
   - **Do not test `withdrawal = 0` by running the loop. Explain it in a comment only.**

### Hint
Your loop condition is `while balance >= withdrawal:`. Make sure something inside the loop changes `balance`, or the loop never ends. If `withdrawal` is `0`, the condition can stay true forever because the balance does not decrease.

````python
balance = 100
withdrawal = 30

# Aabha Kc
# Withdrawal simulator using a while loop

balance = 100
withdrawal = 30

while balance >= withdrawal:
    balance -= withdrawal
    print(f"Withdrew 30, balance is now {balance}")

print("Leftover balance:", balance)

# If withdrawal were 0, the loop would never reduce the balance,
# so the condition would stay True forever and create an infinite loop.

````

## Problem 4 ‚Äî Day-Type Router (switch-case style)  *(10 points)*

Convert a day abbreviation into a message, using **either** `match-case` **or** a **dictionary dispatch** (your choice ‚Äî both are in the lecture).

### Rules
| Code | Message |
|------|---------|
| `"Mon"`, `"Tue"`, `"Wed"`, `"Thu"`, `"Fri"` | `"Weekday ‚Äî time to study!"` |
| `"Sat"`, `"Sun"` | `"Weekend ‚Äî take a break!"` |
| anything else | `"Not a valid day code"` |

### What to do
1. Starter variable: `day = "Sat"`.
2. Set a variable `message` using your chosen approach, then print it.
3. **Test every category:** loop a `for` over `["Mon", "Sat", "Fri", "Sun", "Xyz"]` and print the message for each, like `Mon -> Weekday ‚Äî time to study!`.

### Hint
With `match-case`, you can list several values in one case: `case "Mon" | "Tue" | "Wed" | "Thu" | "Fri":`. With a dictionary, map each code to its message and use `.get(day, "Not a valid day code")`.

````python
day = "Sat"
# Aabha Kc
# Day-type router using match-case style

def get_message(day):
    match day:
        case "Mon" | "Tue" | "Wed" | "Thu" | "Fri":
            return "Weekday ‚Äî time to study!"
        case "Sat" | "Sun":
            return "Weekend ‚Äî take a break!"
        case _:
            return "Not a valid day code"

# Starter variable
day = "Sat"
message = get_message(day)
print(message)

# Testing all cases
for d in ["Mon", "Sat", "Fri", "Sun", "Xyz"]:
    print(f"{d} -> {get_message(d)}")

````

## Problem 5 ‚Äî Safe Average Calculator  *(10 points)*

You're given a list of values that *should* be test scores, but some entries are messy. Compute the average of only the valid ones, **without crashing**.

### What to do
The starter list is provided:
```python
raw_values = ["88", "72", "ninety", "95", "", "60", "100"]
```
1. Loop through `raw_values`. Use a `try` / `except ValueError` block to convert each to an `int`.
   - Valid numbers go into a list `valid`.
   - For invalid entries, print a message like `Skipping invalid value: 'ninety'`.
   - For this assignment, a value is **valid** if it can be converted to an integer. You do **not** need to check whether the score is between 0 and 100.
2. After the loop, **check whether `valid` is empty**:
   - If it has numbers, print the count and the average (rounded to 2 decimals).
   - If it's empty, print `No valid values to average.`
3. Add a **comment**: name one Lesson-4 exception (besides `ValueError`) and one situation that would raise it.

### Hint
This mirrors the lecture's *Safe Score Processor*. `int("ninety")` and `int("")` both raise `ValueError`.

````python
raw_values = ["88", "72", "ninety", "95", "", "60", "100"]

# Aabha Kc
# Safe average calculator using try/except

raw_values = ["88", "72", "ninety", "95", "", "60", "100"]

valid = []

for value in raw_values:
    try:
        num = int(value)
        valid.append(num)
    except ValueError:
        print(f"Skipping invalid value: '{value}'")

# Final check + average
if valid:
    average = sum(valid) / len(valid)
    print("Count:", len(valid))
    print("Average:", round(average, 2))
else:
    print("No valid values to average")

# Example of another exception:
# ZeroDivisionError can occur if you try to divide by zero,
# such as calculating an average when a list is empty.

````
